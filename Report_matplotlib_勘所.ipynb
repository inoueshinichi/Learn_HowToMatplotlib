```python
import numpy as np
import matplotlib.pyplot as plt
```

# 複数のグラフを重ねて表示

```python
x = np.linspace(-5, 5, 300)
y1 = x**2
y2 = (x - 2)**2
```

```python
plt.plot(x, y1, color='r')
plt.plot(x, y2, color='k', linestyle="--")
plt.show()
```

# 複数のグラフを並べて表示する

```python
x = np.linspace(-5, 5, 300)
sin_x = np.sin(x)
cos_x = np.cos(x)
```

```python
fig, axes = plt.subplots(2, 1) # axesはmatplotlib.axes.Axesのインスタンスのリスト
axes[0].set_ylim([-1.5, 1.5])
axes[1].set_ylim([-1.5, 1.5])
axes[0].plot(x, sin_x, color='r')
axes[1].plot(x, cos_x, color='b')
plt.show()
```

## サブプロットが1つしかない場合でも、subplots関数を使用するとサブプロットの違いを吸収できて便利

```python
x = np.linspace(-5, 5, 300)
y = x**2

fig, ax = plt.subplots() # 1つのサブプロット
ax.plot(x, y, color='g')
plt.show()
```

# 等高線の描画

x^2 + y^2/4 = kを満たす曲線を描画

```python
# k=1,2,3,4,5のときの曲線を描画してみる
def f(x, y):
    return x**2 + y**2 / 4

x = np.linspace(-5, 5, 300)
y = np.linspace(-5, 5, 300)

xmesh, ymesh = np.meshgrid(x, y)
z = f(xmesh.ravel(), ymesh.ravel()).reshape(xmesh.shape)

fig, ax = plt.subplots()
ax.contour(x, y, z, levels=[1,2,3,4,5])
plt.show()
```

領域を塗り分ける contourfメソッド
k=1~2の間では、"0.1"で塗りつぶされる
白(1.0) 黒(0.0)

```python
colors = ["0.1", "0.3", "0.5", "0.7"]
levels = [1, 2, 3, 4, 5]
fig, ax = plt.subplots()
ax.contourf(x, y, z, colors=colors, levels=levels)
plt.show()
```
